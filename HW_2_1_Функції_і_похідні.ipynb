1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.

```python
def f(x):
  return 10*x - 5
```

```python
x = np.linspace(0,10,20)

plt.plot(x, f(x), 'o')
```

```python
derivative_vals = (f(x + 0.1) - f(x)) / 0.1
```

```python
plt.plot(x, derivative_vals, 'o')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label="f(x) = 10x - 5", color="blue")
plt.plot(x, derivative_vals, label="f'(x) ≈ 10 (похідна)", color="red")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Графік функції f(x) = 10x - 5 та її похідної")
plt.xlabel("x (години вивчення)")
plt.ylabel("y (оцінка)")
plt.legend()
plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.

### На Python

```python
def numerical_derivative(func, x, h):
    return (func(x + h) - func(x)) / h
```

```python
def f_sin(x):
    return np.sin(x)
```

```python
x_vals = np.linspace(-10, 10, 100)

f_sin_vals = f_sin(x_vals)
f_sin_derivative_vals = numerical_derivative(f_sin, x_vals, h=0.1)


plt.figure(figsize=(10, 6))

plt.plot(x_vals, f_sin_vals, label="f(x) = sin(x)", color="blue")
plt.plot(x_vals, f_sin_derivative_vals, label="f'(x) = cos(x)", color="red")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Графік функції f(x) = sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def f_log(x):
    return np.log(x)
```

```python
# Діапазон значень для x (x > 0, оскільки логарифм визначений тільки для додатних x)
x_vals = np.linspace(0.1, 10, 100)

f_log_vals = f_log(x_vals)
f_log_derivative_vals = numerical_derivative(f_log, x_vals, h=0.1)


plt.figure(figsize=(10, 6))

plt.plot(x_vals, f_log_vals, label="f(x) = ln(x)", color="blue")
plt.plot(x_vals, f_log_derivative_vals, label="f'(x) = 1/x", color="red")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Графік функції f(x) = ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
```

### Використовуючи SymPy

```python
# 𝑓(𝑥)=sin(𝑥)
x = sp.symbols('x')

f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)
derivative_f_sin
```

```python
x_vals = np.linspace(-10, 10, 500)

f_vals = [f_sin.subs(x, val) for val in x_vals]
derivative_vals = [derivative_f_sin.subs(x, val) for val in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label="f(x) = sin(x)", color="blue")
plt.plot(x_vals, derivative_vals, label="f'(x) = cos(x)", color="red")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Графік функції f(x) = sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 𝑓(𝑥)=ln(𝑥)
x = sp.symbols('x')

f_log = sp.ln(x)
derivative_f_log = sp.diff(f_log, x)
derivative_f_log
```

```python
x_vals = np.linspace(0.1, 10, 100)

f_vals = [f_log.subs(x, val) for val in x_vals]
derivative_vals = [derivative_f_log.subs(x, val) for val in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label="f(x) = ln(x)", color="blue")
plt.plot(x_vals, derivative_vals, label="f'(x) = 1/x", color="red")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Графік функції f(x) = ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1
derivative_f = sp.diff(f, x)

print("Похідна функції:", derivative_f)
```

```python
x_vals = np.linspace(0, 100, 500)
f_vals = [f.subs(x, val) for val in x_vals]
derivative_vals = [derivative_f.subs(x, val) for val in x_vals]


plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label="f(x) = x^2 - 5x + 1", color="blue")
plt.plot(x_vals, derivative_vals, label="f'(x) = 2x - 5", color="red")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Графік функції f(x) = x^2 - 5x + 1 та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
